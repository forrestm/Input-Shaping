### Define the Variables

```python
p1, p2, p1s, p2s = sympy.symbols('p1, p2, p1s, p2s')
theta1, theta2, R1, R2 = sympy.symbols('theta1 theta2 R1 R2')
C, z = sympy.symbols('C z')
```

### Define the natural frequencies (Hz) and damping ratios

```python
t = 0.1

R = np.array([0])
theta = np.array([0])
x = np.array([0])
y = np.array([0])
xs = np.array([0])
ys = np.array([0])

nat_freq = np.array([[1],
                     [4]])
damping = np.array([[0.1],
                    [0.1]])

for i in range(np.shape(nat_freq)[0]):
    R_add = np.e**(-(nat_freq[i]*2*damping[i])*np.pi*t)
    theta_add = np.sqrt(1-damping[i]**2) * 2 * nat_freq[i] * np.pi * t
    np.vstack((R,R_add))
    np.vstack((theta,theta_add))
    x_add = R_add*np.cos(theta_add)
    xs_add = R_add*np.cos(-theta_add)
    y_add = R_add*np.sin(theta_add)
    ys_add = R_add*np.sin(-theta_add)
    x = np.vstack((x,x_add))
    xs = np.vstack((xs,xs_add))
    y = np.vstack((y,y_add))
    ys = np.vstack((ys,ys_add))
R = np.delete(R, (0), axis=0)
theta = np.delete(theta, (0), axis=0)
x = np.delete(x, (0), axis=0)
y = np.delete(y, (0), axis=0)
xs = np.delete(xs, (0), axis=0)
ys = np.delete(ys, (0), axis=0)
    

# R_1 = np.e**(-0.2*np.pi*t)
# R_2 = np.e**(-0.8*np.pi*t)

# theta_1 = 1.9900*np.pi*t
# theta_2 = 7.9599*np.pi*t

# x_1 = R_1*np.cos(theta_1)
# y_1 = R_1*np.sin(theta_1)
# x_1s = R_1*np.cos(-theta_1)
# y_1s = R_1*np.sin(-theta_1)

# x_2 = R_2*np.cos(theta_2)
# y_2 = R_2*np.sin(theta_2)
# x_2s = R_2*np.cos(-theta_2)
# y_2s = R_2*np.sin(-theta_2)
```

```python
p1 = R1 * sympy.exp(sympy.I*theta1)
p1s = R1 * sympy.exp(-sympy.I*theta1)

p2 = R2 * sympy.exp(sympy.I*theta2)
p2s = R2 * sympy.exp(-sympy.I*theta2)
```

```python
H = (((z-p1)*(z-p1s)*(z-p2)*(z-p2s)))
sympy.collect(sympy.expand(H), z)
```

```python
t = np.linspace(0,1,10000)
R_1 = np.e**(-0.2*np.pi*t)
R_2 = np.e**(-0.8*np.pi*t)

theta_1 = 1.9900*np.pi*t
theta_2 = 7.9599*np.pi*t

a1 = -2*(R_1 * np.cos(theta_1)+ R_2*np.cos(theta_2))
a2 = (R_1**2) + 4*R_1*R_2*np.cos(theta_1)*np.cos(theta_2)+(R_2**2)
a3 = -2*(R_1*(R_2**2) * np.cos(theta_1)+ (R_1**2)*R_2*np.cos(theta_2))
a4 = (R_1**2) * (R_2**2)

theta_1 = 1.9900*np.pi*t
theta_2 = 7.9599*np.pi*t
plt.plot(t, a1)
plt.plot(t, a2)
plt.plot(t, a3)
plt.plot(t, a4)
plt.xlim(0,0.25)
plt.ylim(-6,6)
```

```python
index = np.argwhere(
np.logical_and(
np.logical_and(np.greater_equal(a1,0),np.greater_equal(a2,0)),
np.logical_and(np.greater_equal(a3,0),np.greater_equal(a4,0))))[0][0]
```

```python
print('A1: {}\nA2: {}\nA3: {}\nA4: {}'.format(a1[index].round(3),
                                              a2[index].round(3),
                                              a3[index].round(3),
                                              a4[index].round(3)))
```

```python
C_sol = float(sympy.solve(C*(1+a1[index]+a2[index]+a3[index]+a4[index])-1,C)[0])
shaper_times = np.array(  [[ 0.0000],
                           [ t[index]],
                           [ t[index]*2],
                           [ t[index]*3],
                           [ t[index]*4]])

shaper_impulses = np.array([[ C_sol  ],
                           [ C_sol*a1[index]],
                           [ C_sol*a2[index]],
                           [ C_sol*a3[index]],
                           [ C_sol*a4[index]]])
Shaper = np.hstack((shaper_times, shaper_impulses))
Shaper
```
